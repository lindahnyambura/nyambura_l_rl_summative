## **Training Analysis**

### **Cumulative Reward plots**

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Define paths to the log files
log_paths = {
    "DQN": "/content/drive/MyDrive/rl_summative_final/logs/dqn_final",
    "PPO": "/content/drive/MyDrive/rl_summative_final/logs/ppo",
    "A2C": "/content/drive/MyDrive/rl_summative_final/logs/a2c",
    "REINFORCE": "/content/drive/MyDrive/rl_summative_final/logs/reinforce",
}

# Function to load cumulative rewards from log files
def load_cumulative_rewards(log_path):
    # Load the evaluations.npz file
    eval_file = os.path.join(log_path, "evaluations.npz")
    if not os.path.exists(eval_file):
        raise FileNotFoundError(f"No evaluations file found at {eval_file}")

    data = np.load(eval_file)
    rewards = data['results'].mean(axis=1)  # Mean reward over episodes
    return np.cumsum(rewards)

# Load cumulative rewards for each method
cumulative_rewards = {method: load_cumulative_rewards(path) for method, path in log_paths.items()}

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

for idx, (method, rewards) in enumerate(cumulative_rewards.items()):
    ax = axs[idx]
    if rewards is not None and len(rewards) > 0:
        ax.plot(rewards, marker='o')
        ax.set_title(f"{method} Cumulative Reward")
        ax.set_xlabel("Evaluation Step")
        ax.set_ylabel("Cumulative Reward")
        ax.grid(True)
    else:
        ax.set_title(f"{method} (No Data)")
        ax.axis('off')

plt.tight_layout()
plt.show()
```

### **Training stability plots**

```python
from tbparse import SummaryReader
import matplotlib.pyplot as plt
import os

log_dirs = {
    "DQN": "/content/drive/MyDrive/rl_summative_final/logs/dqn_final",
    "PPO": "/content/drive/MyDrive/rl_summative_final/logs/ppo",
    "A2C": "/content/drive/MyDrive/rl_summative_final/logs/a2c",
    "REINFORCE": "/content/drive/MyDrive/rl_summative_final/logs/reinforce",
}

# Read tensorboard logs
readers = {k: SummaryReader(os.path.join(v, "tensorboard")) for k, v in log_dirs.items()}

# Plot DQN loss
dqn_df = readers["DQN"].scalars
dqn_loss = dqn_df[dqn_df["tag"].str.contains("loss")]

plt.figure(figsize=(10, 4))
for tag in dqn_loss["tag"].unique():
    df = dqn_loss[dqn_loss["tag"] == tag]
    plt.plot(df["step"], df["value"], label=tag)
plt.title("DQN Training Loss (Objective Function)")
plt.xlabel("Timesteps")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Policy Entropy for PPO, A2C, REINFORCE
plt.figure(figsize=(14, 6))

for i, method in enumerate(["PPO", "A2C", "REINFORCE"]):
    df = readers[method].scalars
    entropy = df[df["tag"].str.contains("entropy")]

    if entropy.empty:
        print(f"[Warning] No entropy data found for {method}")
        continue

    plt.plot(entropy["step"], entropy["value"], label=method)

plt.title("Policy Entropy During Training (Stability Indicator)")
plt.xlabel("Timesteps")
plt.ylabel("Entropy")
plt.legend()
plt.grid(True)
plt.show()
```
